Now lets just suppose we know about four people, we know that they have read the book and watched the movies or not and whether they are potterheads or not. Because we know only people who have done both are **potterheads** (not potheads).
Here for a person we have [a,b] where a and b stand for book and movie respectively, 0 means NO and 1 means YES

```python
#generate input_features and print its shape
input_features = np.array([[0,0],[0,1],[1,0],[1,1]])
print(input_features.shape)
```

Now create a target vector for the examples chosen above with two types of values 0 and 1(no and yes). 

0 - not a Potterhead

1- Potterhead

```python
#define target array
target_output = np.array([[0,0,0,1]])

#convert it into vector of appropriate dimensions
# if input has mXn dimensions, then target will be of nX1
target_output = target_output.reshape(4,1);

print(target_output.shape)
target_output
```

Create a weight matrix for the number of features taken. Values can be any random numbers.

```python
#define weights array
weights = np.array([[0.1],[0.2]])

#print the shape of weights array chosen above
print(weights.shape)
```

```python
#assign suitable bias
b = 0.4

#assign learning rate
lr = 0.1

#define sigmoid function (1/(1+e^(-x)))
def sigmoid(x):
  return (1/(1 + np.exp(-x)))

#Now lets define the key to the gradient descent algorithm: Sigmoid derivative:)
def sigmoid_derivative(x):
  s = sigmoid(x)
  ds = s*(1-s)
  return (ds)
```

Here comes the main logic behind the neural network! Lets run this network 1000 times. The number sounds big but don't worry, it will run in the wink of an eye:))

```python
#define a loop for epochs=1000

for epoch in range(1000):
  inputs = input_features
  #feed forward logic for the network 
  #dot input and weight and add bias
  in_frwd = np.dot(inputs,weights) + b

  #feed forward the output = sigmoid of in_fwd
  out_frwd = sigmoid(in_frwd)

  #print(out_frwd.shape, in_frwd.shape, target_output.shape)

  #Its time for backpropagation


  #Calculate error =bout_frwd - target_output
  error = out_frwd - target_output

  #Sum out the errors for all examples
  s = np.sum(error)

  #Lets begin the gradient descent.........

  #Calculate the derivative
  #calcuate the value adjust that is to be used to correct our weights
  #adjust = error*sigmoid_derivative(out_frwd)
  adjust = error * sigmoid_derivative(out_frwd)

  #Now its time to use our lr.Adjust the weights and don't forget to transpose the input matrix while finding final_derivative
  inputs = input_features.T



  #final_deriavtive = np.dot of inpts and adjust
  final_derivative = (1/4)*np.dot(inputs,adjust)


  # relationship between weights lr anf final derivative?
  weights-= lr * final_derivative

  #Update the bias value for each example
  for i in adjust:
    b-=lr*i

    print(weights," ",b)  #check weight and bias values with each iteration
```

**So, techies its time to predict values using the learned weights and our own inputs with unknown outputs**

1. Here we assume that the person likes wizarding world but does not love Harry much.

```python
#define a single example for which we want to predict considering the no of features we chose prior to training
my_input = np.array([1,0])

#Logic for calculation of output from a neuron using my_input
z = np.sum(np.dot(my_input,weights) + b)
a = sigmoid(z)

print(a)
#How would you declare a person a Potterhead or not based on this wierd output we got above? Lets code!
if a>0.5:
  print("PotterHead!")
else:
  print("Not a PotterHead")
```
